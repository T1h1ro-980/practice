# 1.1 Создание тензоров
Создайте следующие тензоры:
- Тензор размером 3x4, заполненный случайными числами от 0 до 1
- Тензор размером 2x3x4, заполненный нулями
- Тензор размером 5x5, заполненный единицами
- Тензор размером 4x4 с числами от 0 до 15 (используйте reshape)

```python
tensor1 = torch.rand(3, 4)
print(f"Тензор размером 3x4, заполненный случайными числами от 0 до 1: \n {tensor1}")

tensor2 = torch.zeros(2, 3, 4)
print(f"Тензор размером 2x3x4, заполненный нулями: \n {tensor2}")

tensor3 = torch.ones(5, 5)
print(f"Тензор размером 5x5, заполненный единицами: \n {tensor3}")

tensor4 = torch.arange(0, 16).reshape(4, 4)
print(f"Тензор размером 4x4 с числами от 0 до 15: \n {tensor4}")
```

# 1.2 Операции с тензорами
Дано: тензор A размером 3x4 и тензор B размером 4x3
Выполните:
- Транспонирование тензора A
- Матричное умножение A и B
- Поэлементное умножение A и транспонированного B
- Вычислите сумму всех элементов тензора A

```python
A = torch.rand(3, 4, dtype = torch.float32)
B = torch.rand(4, 3, dtype = torch.float32)

a_T = A.T
print(f"Транспонирование тензора A: \n {a_T}")

matrix_multiplication = A @ B
print(f"Матричное умножение A и B: \n {matrix_multiplication}")

matrix_multiplication_2 = A * B.T
print(f"Поэлементное умножение A и транспонированного B: \n {matrix_multiplication_2}")
```

# 1.3 Индексация и срезы
Создайте тензор размером 5x5x5 \
Извлеките:
- Первую строку
- Последний столбец
- Подматрицу размером 2x2 из центра тензора
- Все элементы с четными индексами

```python
tensor = torch.rand(5, 5, 5)
print(f'tensor = \n {tensor}')

first_row = tensor[:1, :, :]
print(f'Первая строка: \n {first_row}')

last_column = tensor[:, :, -1:]
print(f'Последний столбец: \n {last_column}')

submatrix = tensor[1:3, 1:3, 1:3]
print(f'Подматрицу размером 2x2 из центра тензора: \n {submatrix}')

even_idx = tensor[::2, ::2, ::2]
print(f'Все элементы с четными индексами: \n {even_idx}')
```

# 1.4 Работа с формами
Создайте тензор размером 24 элемента \
Преобразуйте его в формы:
- 2x12
- 3x8
- 4x6
- 2x3x4
- 2x2x2x3

```python
tensor = torch.rand(24)
print(f"Тензор 1x24: \n{tensor}")

tensor_2_12 = tensor.reshape(2, 12)
print(f"Тензор 2x12: \n{tensor_2_12}")

tensor_3_8 = tensor.reshape(3, 8)
print(f"Тензор 3x8: \n{tensor_3_8}")

tensor_4_6 = tensor.reshape(4, 6)
print(f"Тензор 4x6: \n{tensor_4_6}")

tensor_2_3_4 = tensor.reshape(2, 3, 4)
print(f"Тензор 2x3x4: \n{tensor_2_3_4}")

tensor_2_2_2_3 = tensor.reshape(2, 2, 2, 3)
print(f"Тензор 2x2x2x3: \n{tensor_2_2_2_3}")
```
